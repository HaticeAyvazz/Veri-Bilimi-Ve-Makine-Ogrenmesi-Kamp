```python
#Sağlık verilerinin çoğunda 0 olma ihtimali yok denecek kadar azdır.Bu kolonların tespiti

columns_to_check=["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]

for col in columns_to_check:
    zero_count=(df[col]==0).sum()
    zero_percentage=100*zero_count/len(df)
    print(f"{col}:{zero_count} %{zero_percentage:.2f}")
```

```python
#independent,dependent feautureların belirlenmesi
X=df.drop("Outcome",axis=1)
y=df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
#Datasette tespit edilen 0 a eşit olan satırların doldurulması
#Burada median değeri sadece train verisinde hesaplandı.Test verisi doldurulurken train de elde edilen median değeriyle dolduruldu.
columns_to_fill=["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]

medians={}

for col in columns_to_fill:
    median_value=X_train[X_train[col]!=0][col].median()
    medians[col]=median_value
    X_train[col]=X_train[col].replace(0,median_value)

for col in columns_to_fill:
    X_test[col]=X_test[col].replace(0,medians[col]) 
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
#Bazı algoritmalar için gereklidir
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

#Model eğitim sonrası hesaplanacak değerlerin fonksiyon aracılığıyla hazırlanması
def calculate_model_metric(true,predicted):
    cr=classification_report(true,predicted)
    cm=confusion_matrix(true,predicted)
    acc=accuracy_score(true,predicted)

    print("Classification Report:\n",cr)
    print("Confusion Matrix:\n",cm)
    print("Accuracy Score:\n",acc)
    print("-----------------------")

    return cr,cm,acc
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
#Model eğitimi yapılacak algoritmalar
models={
    "Logistic Regression":LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
    "SVC":SVC(),
    "AdaBoosting":AdaBoostClassifier()
}
```

```python
for name, model in models.items():

    print("Model:", name)

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    calculate_model_metric(y_test, y_pred)
```

```python
#hyperparameter tuning (Logistic Regression ve AdaBoost için)

log_params = {
    "penalty": ['l1', 'l2'],
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "solver": ['liblinear'],
    "max_iter": [100, 200, 500]
}

ada_params= {
    "n_estimators": [50,70,100,120,150,200],
    "learning_rate":[0.001,0.01,0.1,1,10]
}
```

```python
param_models= [
    ("Log",LogisticRegression(),log_params),
    ("AdaBoost",AdaBoostClassifier(),ada_params)
]
```

```python
from sklearn.model_selection import GridSearchCV

for name,model,params in param_models:
    grid=GridSearchCV(estimator=model,param_grid=params,n_jobs=-1,cv=5)
    grid.fit(X_train,y_train)

    print("Best params for:",name,grid.best_params_)
```

```python
log_reg=LogisticRegression(C=1,max_iter=100,penalty="l1",solver="liblinear")
log_reg.fit(X_train,y_train)

log_reg_pred=log_reg.predict(X_test)

print("Accuracy Score:",accuracy_score(y_test,log_reg_pred))
```

```python
ada_boost=AdaBoostClassifier(learning_rate=1,n_estimators=150)
ada_boost.fit(X_train,y_train)

ada_pred=ada_boost.predict(X_test)

print("Accuracy Score:",accuracy_score(y_test,ada_pred))
```
